## Importing Libraries

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Change the default image format to a vector format
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')
```

## Scatter Plot (Basic)

```python
# creating data
nPoints = 200
x = np.random.rand(nPoints)
y = np.random.rand(nPoints)
```

```python
# Scatter plot => following the Explicit Method
fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(x,y,s=150,c="grey",alpha=0.6,edgecolors="black")
ax.set_xlabel("Variable 1", fontsize=14)
ax.set_ylabel("Variable 2", fontsize=14)
plt.show()
```

## Marker Types

* Matplotlib Markers: https://matplotlib.org/stable/api/markers_api.html
* Latex Symbols: https://matplotlib.org/tutorials/text/mathtext.html

```python
# Scatter plot => following the Explicit Method
fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(x,y,s=250,c="green",alpha=0.6,edgecolors="black",marker=r'$ \pi $')
ax.set_xlabel("Variable 1", fontsize=14)
ax.set_ylabel("Variable 2", fontsize=14)
plt.show()
```

# Adding an Additional Variable $\rightarrow$ Modify Marker Color

<img src="_images/feature_color.PNG" width="900" height="400" align="center"  />

```python
# data
n = 100
x = np.linspace(-2,2,n)
y = x**2 + np.random.rand(n)

# additional variable
z = x**4

# scatter plot
fig = plt.figure()
ax = fig.add_subplot()
im = ax.scatter(x, y, s=150, c=z, alpha=0.5, edgecolors="black",cmap="jet_r")
cbar = fig.colorbar(im)
cbar.set_label("third variable", fontsize=14)
ax.set_xlabel("Variable 1", fontsize=14)
ax.set_ylabel("Variable 2", fontsize=14)
ax.grid(alpha=0.25)
plt.show()
```

# Adding an Additional Variable $\rightarrow$ Marker Size (Bubble Plot)

<img src="_images/feature_bubble.PNG" width="900" height="400" align="center"  />

```python
# data
n = 100
x = np.linspace(0,1,n) 
y = 3*x + np.random.rand(n)

# additional variable
z = x**2

# scatter plot
fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(x, y, s=500*z, c=z, alpha=0.5, edgecolors="black")
ax.set_xlabel("Variable 1", fontsize=14)
ax.set_ylabel("Variable 2", fontsize=14)
ax.grid(alpha=0.25)
plt.show()
```
